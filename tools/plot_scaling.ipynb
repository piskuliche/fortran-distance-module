```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n, cl1, db1 = np.genfromtxt("../rc_3.0.dat", usecols=(0,1,2), unpack=True)
_, cl2, db2 = np.genfromtxt("../rc_6.0.dat", usecols=(0,1,2), unpack=True)
_, cl3, db3 = np.genfromtxt("../rc_12.0.dat", usecols=(0,1,2), unpack=True)
```

```python
fig = plt.figure()
plt.plot(n, cl1, marker='o', c='black', label="3.0 CL")
plt.plot(n, cl2, marker='o',c='red', label="6.0 CL")
plt.plot(n, cl3, marker='o', c='blue', label="12.0 CL")
plt.plot(n, db1, marker='o', c='black',linestyle='dashed', label="3.0  DB")
plt.plot(n, db2, marker='o',c='red',linestyle='dashed', label="6.0 DB")
plt.plot(n, db3, marker='o', c='blue',linestyle='dashed', label="12.0  DB")
plt.xlabel("Number of atoms")
plt.ylabel("Calculation Time")
plt.legend()
plt.show()
```
